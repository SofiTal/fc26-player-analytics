```python
# Vérifier la colonne
print(df['Position'].unique())

# --- Séparation ---

# 1) Gardiens
df_gk = df[df['Position'] == "GK"].copy()

# 2) Joueurs de champ (tout sauf GK)
df_field = df[df['Position'] != "GK"].copy()

# --- Sauvegarde ---
df_gk.to_csv("fc26_gardiens.csv", index=False)
df_field.to_csv("fc26_joueurs_de_champ.csv", index=False)

print("✓ Gardien dataset :", df_gk.shape)
print("✓ Joueurs de champ dataset :", df_field.shape)
```

```python
# Pour le dataset des gardiens
missing_gk = df_gk.isna().sum()
missing_gk_percent = (missing_gk / len(df_gk)) * 100

print("=== Gardiens ===")
print(missing_gk[missing_gk > 0].sort_values(ascending=False))
print(f"Total valeurs manquantes : {missing_gk.sum()}")
print(f"Pourcentage de valeurs manquantes par colonne :\n{missing_gk_percent[missing_gk_percent > 0]}")

# Pour le dataset des joueurs de champ
missing_field = df_field.isna().sum()
missing_field_percent = (missing_field / len(df_field)) * 100

print("\n=== Joueurs de champ ===")
print(missing_field[missing_field > 0].sort_values(ascending=False))
print(f"Total valeurs manquantes : {missing_field.sum()}")
print(f"Pourcentage de valeurs manquantes par colonne :\n{missing_field_percent[missing_field_percent > 0]}")
```

```python
# Pour les gardiens : supprimer 'Alternative positions'
df_gk_clean = df_gk.drop(columns=['Alternative positions'])

# Pour les joueurs de champ : supprimer les colonnes de stats gardien
gk_cols = ['GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes']
df_field_clean = df_field.drop(columns=gk_cols)

# Vérification des valeurs manquantes après nettoyage
print("=== Gardiens ===")
print(df_gk_clean.isna().sum().sum())
print("\n=== Joueurs de champ ===")
print(df_field_clean.isna().sum().sum())
```
